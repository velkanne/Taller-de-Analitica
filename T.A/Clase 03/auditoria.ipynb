```python
import pandas as pd
import numpy as np

# Creamos un dataset con fallas de calidad intencionales
data = {
    'ID_Usuario': [1, 2, 2, 3, 4, 5, 6, 7],
    'Nombre': ['Juan Gomez', 'Ana Silva', 'Ana Silva', 'Pedro Perez', 'Luis Jara', 'Marta Diaz', 'Jose Perez', 'Raul V.'],
    'Edad': [25, 30, 30, -5, 150, 28, np.nan, 35], # Errores de Exactitud y Nulos
    'Email': ['juan@gmail.com', 'ana@cl', 'ana@cl', 'pperez@gmail.com', 'l.jara', 'marta@gmail.com', 'jp@gmail.com', 'raul@gmail.com'], # Errores de Validez
    'Fecha_Registro': ['2024-01-01', '2024-01-01', '2024-01-01', '01/01/24', '2024-01-01', '2023-12-01', '2024-01-01', '2024-01-01'] # Consistencia
}

df_calidad = pd.DataFrame(data)
df_calidad.to_csv('auditoria_calidad.csv', index=False)
print("Archivo 'auditoria_calidad.csv' creado para auditoria!")
```

```python
# Cargar para auditar
df = pd.read_csv('auditoria_calidad.csv')

print("--- REPORTE DE CALIDAD DE DATOS ---")

# 1. Chequeo de UNICIDAD
duplicados = df.duplicated().sum()
print(f"1. Unicidad: Se encontraron {duplicados} registros duplicados.")

# 2. Chequeo de COMPLETITUD
nulos = df.isnull().sum().sum()
print(f"2. Completitud: Faltan {nulos} datos en el dataset.")

# 3. Chequeo de EXACTITUD (Rango de edad lógico: 0 a 120 años)
exactitud = df[(df['Edad'] < 0) | (df['Edad'] > 120)]
print(f"3. Exactitud: Hay {len(exactitud)} registros con edades fuera de rango lógico.")

# 4. Chequeo de VALIDEZ (Formato de email simple: debe contener '@')
validez = df[~df['Email'].str.contains('@')]
print(f"4. Validez: Hay {len(validez)} correos con formato inválido.")

# 5. Chequeo de CONSISTENCIA (Formatos de fecha)
print(f"5. Consistencia: Valores únicos en Fecha:\n{df['Fecha_Registro'].unique()}")
```
