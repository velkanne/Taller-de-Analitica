```python
import pandas as pd
```

```python
fogon_data = {
    'Ticket': ['A1', 'A2', 'A1', 'A4', 'A5', 'A6', 'A7', 'A5'],
    'Fecha_Hora': ['09-03 12:00', '09-03 12:15', '09-03 12:00', '09-03 13:00', '09-03 13:10', '09-03 13:20', '09-03 14:00', '09-03 13:10'],
    'Garzon': ['Wilson', 'Ana', 'Wilson', 'WILSON', 'ana', 'Pedro', None, 'ana'],
    'Monto_CLP': [35000, -5000, 35000, None, 12000, 0, 45000, 12000],
    'Depto': ['Terraza', 'Salon', 'Terraza', 'TERRAZA', 'salon', 'Salon', 'Terraza', 'salon']
}

pd.DataFrame(fogon_data).to_csv('fogon.csv', index=False)
print("Archivo 'fogon.csv' creado.\n")
```

```python
df = pd.read_csv("fogon.csv")

print("1. Total de Registros")
print(f"Filas: {df.shape[0]}\n")

print("2. Detección de Duplicados")
print(f"Total filas duplicadas: {df.duplicated().sum()}\n")

print("3. Detección de Nulos")
print(f"{df.isnull().sum()}\n")

print("4. Detección de Inconsistencias de Texto (Garzón)")
print(f"{df['Garzon'].unique()}\n")

print("5. Detección de Inconsistencias de Texto (Depto)")
print(f"{df['Depto'].unique()}\n")

print("6. Detección de Valores Anómalos (Monto_CLP <= 0)")
print(df[df['Monto_CLP'] <= 0])
```
